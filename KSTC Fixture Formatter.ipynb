```python
import pandas as pd
```

```python
data = pd.read_csv('./KSTC Winter 2022-23 Fixtures.csv')
data.head()
```

```python
output = ""
for i in range(0, len(data)):
    if(data.iloc[i,2] == "Home"):
        venue = "Home Match"
        ks = 6
        opp = 7
    elif(data.iloc[i,2] == "Home (Bicester)"):
        venue = "Home Match (at Bicester)"
        ks = 6
        opp = 7
    elif(data.iloc[i,2] == "Home (Deddington)"):
        venue = "Home Match (at Deddington)"
        ks = 6
        opp = 7
    else:
        venue = "Away Match"
        ks = 7
        opp = 6
     
    date = data.iloc[i,0].split()
    
    if((date[1] == "1") or (date[1] == "21") or (date[1] == "31")): 
        sup = "st"
    elif((date[1] == "3") or (date[1] == "23")):
        sup = "rd"
    elif((date[1] == "2") or (date[1] == "22")):
        sup = "nd"
    else:
        sup = "th"

    team = data.iloc[i,4].split()

    output += "<p><strong>" +  date[0] + " " + date[1] + "<sup>" + sup + "</sup> " + date[2] + "</strong> at " + data.iloc[i,1] + " - " + venue + "<br />\n"
    output += data.iloc[i,ks] + " " + team[0] + " vs. " + data.iloc[i,opp] + "<br />\n"
    output += data.iloc[i,3] + " " + data.iloc[i,4] + " " + data.iloc[0,5] + "<br /></p>\n\n"

print(output)
```
